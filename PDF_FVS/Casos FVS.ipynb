```python
def insere_data(df,data):
    df['Data'] = data
    df['Data'] = pd.to_datetime(df['Data'].astype(str), format='%d-%m-%Y').dt.date
    
    return df

def geraDataframe(caminho,data):
    dia = int(data[0]+data[1])
    mes = int(data[3]+data[4])
    
    #Alguns arquivos precisam de ajustes especificos
    if(mes==6):
        if(dia<=5):
            df = tabula.read_pdf(caminho,pages ='1',area=(102.57974910736084,54.18314952850342,566.5558450698853,518.1592454910278),
                        lattice=True,pandas_options={'header':None})
            df = df[0]
            df = df.dropna(axis=1, how='all')
            df = df.dropna(axis=0)
            df.columns = ['Municipio','Notificados','Confirmados','Clinico Epidemiologico','Participacao por Municipio','Incidencia',
                 'Obitos','Letalidade','Mortalidade']
            
            df = insere_data(df,data)
            return df
        elif(dia==6):
            df = tabula.read_pdf(caminho,pages ='1',area=(98.39939800262452,53.14619892120362,551.9837887954712,515.1497895431519),
                        lattice=True,pandas_options={'header':None})
            df = df[0]
            df = df.dropna(axis=1, how='all')
            df = df.dropna(axis=0)
            df.columns = ['Municipio','Notificados','Confirmados','Clinico Epidemiologico','Participacao por Municipio','Incidencia',
                 'Obitos','Letalidade','Mortalidade']
            
            df = insere_data(df,data)
            return df
        elif(dia==7):
            df = tabula.read_pdf(caminho,pages ='1',area=(211.62919990539552,106.3744655609131,1138.7666471099853,1091.7379360198975),
                        stream=True,pandas_options={'header':None})
            df = df[0]
            df = df.dropna(axis=1, how='all')
            df.columns = ['Municipio','Notificados','Confirmados','Clinico Epidemiologico','Participacao por Municipio','Incidencia',
                 'Obitos','Letalidade','Mortalidade']
            
            df = insere_data(df,data)
            return df
        elif(dia>=8 and dia<=14):
            if(dia==12):
                df = tabula.read_pdf(caminho,pages ='1',area=(118.36124897003174,49.02684955596924,562.3474451065064,513.950845527649),
                        stream=True,pandas_options={'header':None})
                df = df[0]
                df = df.dropna(axis=1, how='all')
                df = df.drop(columns=[1])
                df[0][0] = 'Manaus'
                df[0][1] = 'INTERIOR'
                df.columns = ['Municipio','Notificados','Confirmados','Clinico Epidemiologico','Participacao por Municipio','Incidencia',
                 'Obitos','Letalidade','Mortalidade']
                
                df = insere_data(df,data)
                return df
            else:
                df = tabula.read_pdf(caminho,pages ='1',area=(118.36124897003174,49.02684955596924,562.3474451065064,513.950845527649),
                        stream=True,pandas_options={'header':None})
                df = df[0]
                df = df.dropna(axis=1, how='all')
                df.columns = ['Municipio','Notificados','Confirmados','Clinico Epidemiologico','Participacao por Municipio','Incidencia',
                 'Obitos','Letalidade','Mortalidade']
                
                df = insere_data(df,data)
                return df
        elif((dia>=16 and dia<=21) or (dia>=26 and dia<=30)):
            df = tabula.read_pdf(caminho,pages ='1',area=(118.36124897003174,49.02684955596924,562.3474451065064,513.950845527649),
                        stream=True,pandas_options={'header':None})
            df = df[0]
            df = df.dropna(axis=1, how='all')
            df = df.drop(columns=[1])
            df[0][0] = 'Manaus'
            df[0][1] = 'INTERIOR'
            df.columns = ['Municipio','Notificados','Confirmados','Clinico Epidemiologico','Participacao por Municipio','Incidencia',
                 'Obitos','Letalidade','Mortalidade']
            
            df = insere_data(df,data)
            return df
        elif(dia>=22 and dia<=25):
            df = tabula.read_pdf(caminho,pages ='1',area=(98.39939800262452,48.936599006652834,549.8789888381958,520.4117894363404),
                        stream=True,pandas_options={'header':None})
            df = df[0]
            df = df.dropna(axis=1, how='all')
            df = df.drop(columns=[1])
            df[0][0] = 'Manaus'
            df[0][1] = 'INTERIOR'
            df.columns = ['Municipio','Notificados','Confirmados','Clinico Epidemiologico','Participacao por Municipio','Incidencia',
                 'Obitos','Letalidade','Mortalidade']
            
            df = insere_data(df,data)
            return df
    elif(mes==7):
        df = tabula.read_pdf(caminho,pages ='1',area=(96.26714916229248,47.87054958343506,558.1390451431274,507.63824558258057),
                        lattice=True,pandas_options={'header':None})
        df = df[0]
        df = df.dropna(axis=1, how='all')
        df = df.dropna(axis=0)
        df.columns = ['Municipio','Notificados','Confirmados','Clinico Epidemiologico','Participacao por Municipio','Incidencia',
                 'Obitos','Letalidade','Mortalidade']
        
        df = insere_data(df,data)
        return df
    elif(mes==8):
        df = tabula.read_pdf(caminho,pages ='1',area=(87.85034923553467,47.87054958343506,558.1390451431274,513.950845527649),
                        lattice=True,pandas_options={'header':None})
        df = df[0]
        df = df.dropna(axis=1, how='all')
        df = df.dropna(axis=0)
        df.columns = ['Municipio','Notificados','Confirmados','Clinico Epidemiologico','Participacao por Municipio','Incidencia',
                 'Obitos','Letalidade','Mortalidade']
        
        df = insere_data(df,data)
        return df

def geraCSV(destino,operacao,df,csv_completo):
    if(operacao == 'n'):
        nome = str(destino+'c.csv')
        df.to_csv(nome)
        print('Finalizada a criação do csv!')
    else:
        #recupera o df completo a ser concatenado
        dfCompleto = pd.read_csv(csv_completo)
        dfFinal = pd.concat([df,dfCompleto], ignore_index = True)
        dfFinal = dfFinal.drop('Unnamed: 0', axis = 1)
        nome = str(csv_completo)
        dfFinal.to_csv(nome)
        print('Finalizada a concatenação do csv!')
```
