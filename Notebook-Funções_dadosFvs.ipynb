```python
import tabula
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
```

```python
#Lendo somente a planilha de casos contidas no pdf 

def leitor_pdf(pdf):
    data = tabula.read_pdf(pdf, area = [90,10,1150,1050], pages ='1')
    
    data = data[0]

    return data

# organizando os dados deslocados e excluíndo as linhas 'manaus','interios','total'
def organiza_pdf(data):
    df1 = data.drop(['Unnamed: 0','Unnamed: 1'], axis = 1)
    df1 = df1.drop([0,1,2,3,4,66])
    return df1

#organizando os dados das linhas excluidas
def organiza_pdf2(data):
    temp = data.loc[3:4]
    temp = temp.drop(['Unnamed: 0','Unnamed: 9'], axis=1)
    
    return temp

#renomeando as colunas
def renome_pdf(dado):
    dado.columns = ['Municipio','Notificados','Confirmados','Clinico Epidemiologico','Participação por Municipio','Incidencia',' Obitos','Letalidade','Mortalidade']
    return dado

# unindo as linhas em um dataframe depois de formatadas
def junta_tudo(dado1, dado2):
    dado1 = dado1.append(dado2)
    dado1 = dado1.sort_index()
    
    return dado1

#inserindo a data da planilha
def insere_data(df,data):
    df['Data'] = data
    df['Data'] = pd.to_datetime(df['Data'])
    
    return df


# salvando com .csv
def salva_df_fvs(dados):
    
    dados.to_csv('dataFrame_fvs.csv')
    
    return dados
    
```
