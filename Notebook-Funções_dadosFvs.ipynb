```python
import tabula
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
```

```python
#Lendo somente a planilha de casos contidas no pdf 
data = tabula.read_pdf('boletim-diario-de-casos-covid-19-06-06-2020_HDR1txL.pdf', area=[90,10,600,600], pages ='1'
)
data = data[0]
```

```python
def organiza_pdf(data):
    df1 = data.drop(['Unnamed: 0','Unnamed: 1'], axis = 1)
    df1 = df1.drop([0,1,2,64])
    return df1

def organiza_pdf2(data):
    temp = data.loc[1:2]
    temp = temp.drop(['Unnamed: 7','Unnamed: 0'], axis=1)
    
    return temp

def renome_pdf(dado):
    dado.columns = ['Municipio','Notificados','Confirmados','Clinico Epidemiologico','% Participação por Municipio','Incidencia','Nº de Obitos','Letalidade','Mortalidade']
    return dado

def junta_tudo(dado1, dado2):
    dado1 = dado1.append(dado2)
    dado1 = dado1.sort_index()
    
    return dado1


dados_1 = organiza_pdf(data)
dados_2 = organiza_pdf2(data)

dados_1 = renome_pdf(dados_1)
dados_2 = renome_pdf(dados_2)

df_fvs = junta_tudo(dados_1,dados_2)

df_fvs


    
```
