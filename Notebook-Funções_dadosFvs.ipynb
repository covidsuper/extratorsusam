```python
import tabula
import pandas as pd
import numpy as np
import glob
```

```python
#Lendo somente a planilha de casos condas no pdf 

def leitor_pdf(pdf):
    data = tabula.read_pdf(pdf, area = [90,10,1150,1050], pages ='1')
    
    data = data[0]

    return data

# organizando os dados deslocados e excluíndo as linhas 'manaus','interios','total'
def organiza_pdf(data):
    df1 = data.drop(['Unnamed: 0','Unnamed: 1'], axis = 1)
    df1 = df1.drop([0,1,2,3,4,66])
    return df1

#organizando os dados das linhas excluidas
def organiza_pdf2(data):
    temp = data.loc[3:4]
    temp = temp.drop(['Unnamed: 0','Unnamed: 9'], axis=1)
    
    return temp

#renomeando as colunas
def renome_pdf(dado):
    dado.columns = ['Municipio','Notificados','Confirmados','Clinico Epidemiologico','Participacao por Municipio','Incidencia',' Obitos','Letalidade','Mortalidade']
    return dado

# unindo as linhas em um dataframe depois de formatadas
def junta_tudo(dado1, dado2):
    dado1 = dado1.append(dado2)
    dado1 = dado1.sort_index()
    
    return dado1

#inserindo a data da planilha
def insere_data(df,data):
    df['Data'] = data
    df['Data'] = pd.to_datetime(df['Data'])
    
    return df


# salvando com .csv
def salva_df_fvs(dados):
    
    dados.to_csv('dataFrame_fvs.csv')
    
    return dados



    

def gera_dfGeral():
    
    extensao = 'csv'


    arquivos = [i for i in glob.glob('arquivos_csv_fvs/*'.format(extensao))]

    df_geral = pd.concat([pd.read_csv(f,sep = ',') for f in arquivos], ignore_index = True)

    df_geral = df_geral.drop('Unnamed: 0', axis = 1)

    return df_geral

    
    
```

```python
#entra com o nome do pdf

pdf = 'boletim-diario-de-casos-covid-19-10-06-2020.pdf'
dados = leitor_pdf(pdf)
dados
```

```python
# fazendo a primeira formatação
dados_1  = organiza_pdf(dados)
dados_2  = organiza_pdf2(dados)

#renomeando os arquivos
dados_1 = renome_pdf(dados_1)
dados_2 = renome_pdf(dados_2)

#juntando em um arquivo
dados = junta_tudo(dados_1,dados_2)

data = str(input("Adicione a data do PDF (mes-dia-ano):"))
dados = insere_data(dados,data)
dados
```

```python
#salvando o arquivo como .csv no diretório

nome = str('arquivos_csv_fvs/' + input('nome:')+'.csv')

dados.to_csv(nome)
```

```python

df_fvs = gera_dfGeral()
df_fvs
```

```python
#salvando os arquivos concatenados como .csv

df_fvs = salva_df_fvs(df_fvs)
```
